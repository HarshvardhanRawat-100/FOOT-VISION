```python
import sys
sys.path.append("..")

import os
os.chdir("..")
print("Now working from:", os.getcwd())

from src.components.data_ingestion import DataIngestion
from src.components.data_transformation import DataTransformation
from src.components.model_trainer import ModelTrainer
from src.components.model_evaluation import ModelEvaluation

train_csv, test_csv = DataIngestion().initiate_data_ingestion()
DataTransformation().initiate_data_transformation(train_csv, test_csv)
model_path, best_acc = ModelTrainer().initiate_model_training()
print(f"Best validation accuracy: {best_acc:.2%}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", font_scale=1.05)
PALETTE = ["#0f4c3a", "#1a7a5e", "#2fa084", "#5cbfa3", "#8ad4bf", "#b8e6d8"]
plt.rcParams["figure.dpi"] = 110
plt.rcParams["axes.edgecolor"] = "#cccccc"
```

confusion matrix heatmap

```python
import json
import numpy as np

report = ModelEvaluation().initiate_model_evaluation()
with open("artifacts/model_evaluation/report.json") as f:
    full_report = json.load(f)

cm = np.array(full_report["confusion_matrix"])
class_names = [k for k in full_report["classification_report"].keys()
               if k not in ("accuracy", "macro avg", "weighted avg")]

fig, ax = plt.subplots(figsize=(7, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Greens", cbar=False,
            xticklabels=class_names, yticklabels=class_names,
            linewidths=1.5, linecolor="white", annot_kws={"fontsize": 13, "fontweight": "bold"}, ax=ax)
ax.set_xlabel("Predicted", fontsize=12, fontweight="bold")
ax.set_ylabel("Actual", fontsize=12, fontweight="bold")
ax.set_title(f"Confusion Matrix вЂ” {best_acc:.0%} Best Validation Accuracy", fontsize=14, fontweight="bold", pad=15)
plt.xticks(rotation=35, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
```

per-class precision/recall bar chart

```python
metrics_df = pd.DataFrame({
    cls: {"precision": full_report["classification_report"][cls]["precision"],
          "recall": full_report["classification_report"][cls]["recall"]}
    for cls in class_names
}).T

fig, ax = plt.subplots(figsize=(9, 5))
metrics_df.plot(kind="bar", ax=ax, color=[PALETTE[0], PALETTE[3]], edgecolor="white", width=0.7)
ax.set_title("Precision & Recall by Shot Class", fontsize=14, fontweight="bold", pad=12)
ax.set_ylabel("Score")
ax.set_ylim(0, 1.15)
ax.set_xticklabels(class_names, rotation=35, ha="right")
ax.spines[["top", "right"]].set_visible(False)
ax.legend(frameon=False)
plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objects as go
from sklearn.decomposition import PCA
from src.utils import load_numpy_array

X_train = load_numpy_array("artifacts/data_transformation/train.npy")
y_train = load_numpy_array("artifacts/data_transformation/train_labels.npy")
X_flat = X_train.reshape(X_train.shape[0], -1)

pca = PCA(n_components=3)
X_3d = pca.fit_transform(X_flat)

fig = go.Figure()

for i, cls in enumerate(class_names):
    mask = y_train == i
    fig.add_trace(go.Scatter3d(
        x=X_3d[mask, 0], y=X_3d[mask, 1], z=X_3d[mask, 2],
        mode="markers",
        name=cls,
        marker=dict(
            size=7,
            color=PALETTE[i % len(PALETTE)],
            line=dict(width=1, color="white"),
            opacity=0.9,
        ),
        hovertemplate=f"<b>{cls}</b><extra></extra>",
    ))

fig.update_layout(
    title=dict(
        text="Shot Classes in Movement Space (PCA of Keypoint Sequences)",
        font=dict(size=18, family="Arial", color="#1a1a1a"),
        x=0.5,
    ),
    scene=dict(
        xaxis_title=f"PC1 ({pca.explained_variance_ratio_[0]:.1%} variance)",
        yaxis_title=f"PC2 ({pca.explained_variance_ratio_[1]:.1%} variance)",
        zaxis_title=f"PC3 ({pca.explained_variance_ratio_[2]:.1%} variance)",
        xaxis=dict(backgroundcolor="rgba(0,0,0,0)", gridcolor="#e0e0e0"),
        yaxis=dict(backgroundcolor="rgba(0,0,0,0)", gridcolor="#e0e0e0"),
        zaxis=dict(backgroundcolor="rgba(0,0,0,0)", gridcolor="#e0e0e0"),
        camera=dict(eye=dict(x=1.6, y=1.6, z=1.1)),
    ),
    legend=dict(
        title="Shot Class",
        font=dict(size=12),
        bgcolor="rgba(255,255,255,0.8)",
    ),
    width=900,
    height=700,
    margin=dict(l=0, r=0, t=60, b=0),
    paper_bgcolor="white",
)

fig.show()
```
